## Extract and Download Sections from 10-K Filings Over Multiple Years

The `multiprocessing` library is used to parallelize this process, spawning four parallel threads, each responsible for extracting all sections of a single filing. The extracted sections are immediately ready for further analysis; if analysis is planned for a later stage, consider saving the sections either locally or in a document database to avoid "out of memory" errors.
